# **Central Limit Theorem**

In probability theory, the central limit theorem (CLT) establishes that, in most situations, when independent random variables are added, their properly normalized sum tends toward a normal distribution (a bell curve) even if the original variables themselves are not normally distributed.

In this task we build a bimodal distribution then we show that distrobution of sample means approximates a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
x_mean=[]
# loop for 10000 sample from the bimodel distribution and mean over them
for _ in range(10000):
  N = 100
  x1 = np.random.normal(25,5,N)      # 1st distribution with mu=25, sigma=5
  x2 = np.random.normal(5,3,N)       # 2nd distribution
  x = np.concatenate((x1,x2))        
  x_mean.append(x.mean())            # calculate the mean the concatintation of x1 and x2



#plot bimodal distrobution
plt.figure(figsize = (15,5))
sns.kdeplot(x)
plt.title("Bimodal Distribution")
plt.show()


# we are expecting that distribution of sample means is a normal distribution
mu=np.mean(x_mean)
sigma=np.std(x_mean)
plt.figure(figsize = (15,5))
count, bins, ignored = plt.hist(x_mean, 40, density=True)   
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * 
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
               linewidth=2, color='r')
plt.title('Distribution of Sample Means')

plt.show()
```

# **Task2**

The goal of this task is to make use of polynomial or nonlinear regression. The workflow is divided according to the following steps:

1.  Nonlinear data generation function:  $a \sin(x^2) + bx$
2.  Adding some random noise to the generated data.
3.  Applying *curve_fit()* function to find best fit of the data.
4. Plotting data points and   

```python
from scipy import optimize
import matplotlib.pyplot as plt

#data genration
def f(x, a, b):
      return  a*np.sin(x**2) + b*x 

x = np.linspace(0, 7, 501)
y = f(x, 2, 0.7) + np.random.random(len(x)) 


# fit the function
alpha = optimize.curve_fit(f, xdata = x, ydata = y)[0]
print(alpha)


plt.figure(figsize = (15,5))
plt.plot(x, y, 'b.')
plt.plot(x,f(x,*alpha), 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
